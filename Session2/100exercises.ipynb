28. Consider a random 10x2 matrix representing cartesian coordinates, convert them to polar coordinates (★★☆)

```haskell
rand <- randIO' [10,2] :: IO Tensor
let x = Torch.select 1 0 rand
let y = Torch.select 1 1 rand
let newX = Torch.sqrt (x*x + y*y)
let newY = Torch.atan (y/x)
let res = Torch.stack (Dim 1) [newX, newY]
res
```

34. How to find the closest value (to a given scalar) in a vector? (★★☆)

```haskell
let vec= asTensor ([1, 4, 2, 8, 4] :: [Int])
let scalar = asTensor (5 :: Int)
let diff = Torch.abs (vec - scalar)
let idx = Torch.argmin diff 0 False
let res = Torch.select 0 (asValue idx :: Int) vec
res
```

36. Consider a random vector with shape (100,2) representing coordinates, find point by point distances (★★☆)

```haskell
rand <- randIO' [100, 2] :: IO Tensor
let x = Torch.select 1 0 rand
let y = Torch.select 1 1 rand
let xCol = Torch.unsqueeze (Dim 1) x
let yCol = Torch.unsqueeze (Dim 1) y
let xRow = Torch.unsqueeze (Dim 0) x
let yRow = Torch.unsqueeze (Dim 0) y
let dx = xCol - xRow
let dy = yCol - yRow
let res = Torch.sqrt (dx*dx + dy*dy)
res
```
